# Calculating IRR

**Formula**

$$NPV = \sum_{t=0}^{n} \frac{C_t}{(1 + IRR)^t} = 0$$

Where we set NPV to 0, use FCFE (Free Cash Flow from Equity) as Cash from at time *t*. 

FCFE will be adjusted by debt levels of each company -> if stable, use FCFF as a proxy, if not, adjust with change in total debt

**FCFE for NKE, ATZ, LULU from Bloomberg**

```python
"""
For Nike, leverage has been broadly stable over the last several years, 
and Bloomberg consensus does not anticipate material debt issuance or deleveraging across the forecast horizon. 
Since net borrowing is effectively zero, FCFF is a reasonable proxy for FCFE. 
We therefore use consensus Free Cash Flow as the equity cash flow stream when computing IRR.
"""
NKE_FCFF_2026 = 2635.2 * 1000000

#since debt is relatively stable, FCFE = FCFF
NKE_FCFE_2026 = NKE_FCFF_2026

print(f"NIKE FCFF: {NKE_FCFF_2026}")
print(f"NIKE FCFE: {NKE_FCFE_2026}")

"""
Aritzia’s rising leverage means consensus FCFF materially understates equity cash flows in the near term. 
We adjust to FCFE using the change in long-term debt to correctly reflect cash available to equity holders. 
This also informs our risk framing — LULU offers higher upside IRR, but that is partially debt-amplified
"""

ATZ_FCFF_2026 = 384.9 * 1000000

ATZ_TOTAL_DEBT_2024 = 1138.6 * 1000000
ATZ_TOTAL_DEBT_2025 = 1361.2 * 1000000
ATZ_DELTA_TOTAL_DEBT = ATZ_TOTAL_DEBT_2025 - ATZ_TOTAL_DEBT_2024

ATZ_FCFE_2026 = ATZ_FCFF_2026 + ATZ_DELTA_TOTAL_DEBT



"""
Lululemon introduced modest long-term borrowing post-2020, but leverage increases since then have been gradual rather than structural. 
Since net borrowing remains limited, FCFF remains an appropriate primary cash flow proxy for IRR, with a supplementary FCFE-adjusted sensitivity to capture leverage effects.
"""
LULU_FCFF_2026 = 1152.6 * 1000000

LULU_TOTAL_DEBT_2024 = 2859.9 * 1000000
LULU_TOTAL_DEBT_2025 = 3279.2 * 1000000
LULU_DELTA_TOTAL_DEBT = LULU_TOTAL_DEBT_2025 - LULU_TOTAL_DEBT_2024

LULU_FCFE_2026 = LULU_FCFF_2026 + LULU_DELTA_TOTAL_DEBT



```
